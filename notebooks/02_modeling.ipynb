Exploratory modeling: RFM K-Means and Market Basket Analysis.

Run `make etl` before executing cells.

```python
import sys
sys.path.insert(0, '../src')
import duckdb
import pandas as pd
from utils.config import get_settings
from features.rfm import compute_rfm, score_rfm
from models.rfm_segmentation import segment_customers
from models.market_basket import build_basket_matrix, run_apriori

conn = duckdb.connect(get_settings().duckdb_path, read_only=True)
```

```python
# RFM Segmentation
rfm = compute_rfm(conn)
scored = score_rfm(rfm)
segmented = segment_customers(scored)
print(segmented[['customer_id','segment','rfm_score']].head(10))
```

```python
# Market Basket Analysis
basket = build_basket_matrix(conn)
rules = run_apriori(basket, min_support=0.02, min_confidence=0.1)
if not rules.empty:
    print(rules.sort_values('lift', ascending=False).head(10))
else:
    print('No rules found at current thresholds')
```
